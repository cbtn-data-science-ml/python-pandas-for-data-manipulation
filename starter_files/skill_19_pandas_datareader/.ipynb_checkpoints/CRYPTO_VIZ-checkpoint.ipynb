### Import libraries

````python
import pandas as pd
import datetime as dt
from pandas_datareader import data

# Visualization libraries
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
````

````python
# Some stats for SOL-
SOL.describe()
````

### Get crypto currency data

````python
data.DataReader(name = 'SOL-USD', data_source = 'yahoo', start = '2022-01-01', end = '2022-01-31')
````

````python
# Start date set to one year back
start = dt.datetime(end.year-1,end.month,end.day)

# Setting the end date to today
end = dt.datetime.today()
````

````python
# Or you can define crypto currency data like this 
# using the variable name: SOL
SOL = data.DataReader('SOL-USD','yahoo',start,end)
````

````python
SOL
````

````python
SOL.info()
````

````python
SOL.describe()
````

````python
# Visualization of adjusted closing prices
import matplotlib.pyplot as plt
SOL['Adj Close'].plot(legend = True)
````

````python
# Plot the total volume being traded over time
SOL['Volume'].plot(legend=True,figsize=(12,5))
````

````python
# Create moving averages for 10, 20 and 50 day periods of time

ma_days = [10,20,50]

for ma in ma_days:
    column_name = "MA %s days" %(str(ma))
    SOL[column_name] = SOL['Adj Close'].rolling(window=ma,center=False).mean()
````

````python
SOL
````

````python
# Moving averages visualized
SOL[['Adj Close','MA 10 days','MA 20 days','MA 50 days']].plot(legend=True,figsize=(12,5))
````

````python
# Define crypto currency data
BTC = data.DataReader('BTC-USD','yahoo',start,end)
ETH = data.DataReader('ETH-USD','yahoo',start,end)
SOL = data.DataReader('SOL-USD','yahoo',start,end)
DODGE = data.DataReader('DOGE-USD','yahoo',start,end)
````

````python
#We are going to analyze the top cryptocurrencies.
cc_list = ['BTC-USD','ETH-USD','SOL-USD','DOGE-USD']
````

### Daily return avereage

````python
#In order calculate daily return we can use the percentage change of the adjusted closing price

SOL['Daily Return'] = SOL['Adj Close'].pct_change()

SOL['Daily Return'].tail()
````

````python
# Plot the daily return
SOL['Daily Return'].plot(marker='.',legend=True,figsize=(12,5))
````

````python
# Read the 'Adj Close' column from all the cryptocurrencies

close_df = data.DataReader(cc_list,'yahoo',start,end)['Adj Close']
````

````python
close_df.tail()
````

````python
# Explore returns percentage change from the adj close.
returns_df = close_df.pct_change()
returns_df
````

````python
returns_df.plot(marker='.',legend=True,figsize=(12,5))
````

````python
# Find the correlation between Bitcoin and Solana
sns.jointplot(x = 'BTC-USD', y = 'SOL-USD', data = returns_df, kind='scatter')
````

````python
sns.pairplot(returns_df.dropna(),diag_kind='kde')
````

````python
corr = returns_df.dropna().corr()

sns.heatmap(corr,
            xticklabels=corr.columns,
            yticklabels=corr.columns, 
            annot=True,
            cmap='Blues')

#the darker the shade the higher the correlation
````

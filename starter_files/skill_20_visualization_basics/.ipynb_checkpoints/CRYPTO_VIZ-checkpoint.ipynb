### Create moving averages for 10, 20 and 50 day periods of time
A moving average (MA) is a widely used technical indicator that smooths out price trends by filtering out the noise from random short-term price fluctuations.

````python
# A moving average is a calculation to analyze data points by creating a 
# series of averages of different subsets of the full data set. 
# It is also called a moving mean or rolling mean.
# Use .rolling() to perfor rolling window calculations.

periods = [10,20,50]

for days in periods:
    column_name = "MA %s days" %(str(days))
    SOL[column_name] = SOL['Adj Close'].rolling(window=days,center=False).mean()
````

````python
# Moving averages visualized
SOL[['Adj Close','MA 10 days','MA 20 days','MA 50 days']].plot(legend=True,figsize=(12,5))
````

### Predict future behavior

````python
days = 365
d = 1/365
mu = returns.mean()['SOL-USD']
sigma = returns.std()['SOL-USD']
````

````python
#Function takes in stock price, number of days to run, mean and standard deviation values
def stock_monte_carlo(start_price, days, mu, sigma):
    
    price = np.zeros(days)
    price[0] = start_price
    
    shock = np.zeros(days)
    drift = np.zeros(days)
    
    for x in range(1,days):
        
        #Shock and drift formulas taken from the Monte Carlo formula
        shock[x] = np.random.normal(loc=mu*d,scale=sigma*np.sqrt(d))
        
        drift[x] = mu * d
        
        #New price = Old price + Old price*(shock+drift)
        price[x] = price[x-1] + (price[x-1] * (drift[x]+shock[x]))
        
    return price
````

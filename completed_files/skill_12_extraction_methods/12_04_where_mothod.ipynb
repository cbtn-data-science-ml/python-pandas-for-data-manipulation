# `.where()` method

### Import pandas

```python
import pandas as pd
```

### Read CSV and set rating index sorted by labels

```python
cereal = pd.read_csv("cereal.csv", index_col="name")
cereal.sort_index(inplace=True)
cereal
```

### Create a subset mask

```python
mask_1 = cereal["mfr"] == "K"
cereal[mask_1]
```

```python
# now using the .where() method
cereal.where(mask_1)
```

```python
## using comparison operators
mask_2 = cereal["rating"] > 50
cereal[mask_2]
```

```python
# now using the .where() method
cereal.where(mask_2)
```

```python
# using multiple
cereal.where(mask_1 & mask_2)
```
